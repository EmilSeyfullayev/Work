### Preprocess

```python
import numpy as np
import pandas as pd
```

```python
def data_preprocesser(data_path: str):
  
  df = pd.read_excel(data_path)
  df = df.drop(['Unnamed: 5',	'Unnamed: 6'], axis=1)
  df = df.iloc[7:].reset_index(drop=True)
  df.columns = ['date', 'time', 'point', 'direction', 'action', 'name']
  df = df[~df['name'].str.contains("Polis turniket")].reset_index(drop=True)

  date_time = []

  for i in range(len(df)):
    date = df['date'][i]
    time = df['time'][i]

    date_time.append(f'{date} {time}')

  df['date_time'] = date_time
  df['date_time'] = pd.to_datetime(df['date_time'])
  df['weekday'] = df['date_time'].apply(lambda x: x.weekday() + 1)

  df['point']  = df['point'].apply(lambda x: x.replace("Turniket Polis 2 (17)", 'main'))

  return df

df = data_preprocesser("/content/drive/MyDrive/Project Security/5.xls")
```

```python
names = list(df['name'].unique())
dates = list(df['date'].unique())
```

```python
ad2 = []
tarix2 = []
in_datetime = []

for name in names:

  dates = df[df['name'] == name]['date'].unique()

  for date in dates:

    if df[
        (df['name'] == name) &
        (df['date'] == date)
      ]['direction'].iloc[0] == "giriş":

      ad2.append(name)
      tarix2.append(date)

      in_datetime.append(df[
        (df['name'] == name) &
        (df['date'] == date)
      ]['date_time'].iloc[0])


    else:

      ad2.append(name)
      tarix2.append(date)

      in_datetime.append(np.NAN)
```

```python
ad = []
tarix = []
out_datetime = []

for name in names:

  dates = df[df['name'] == name]['date'].unique()

  for date in dates:

    if df[
        (df['name'] == name) &
        (df['date'] == date)
      ]['direction'].iloc[-1] == "çıxış ":

      ad.append(name)
      tarix.append(date)

      out_datetime.append(df[
        (df['name'] == name) &
        (df['date'] == date)
      ]['date_time'].iloc[-1])


    else:

      ad.append(name)
      tarix.append(date)

      out_datetime.append(np.NAN)
```
